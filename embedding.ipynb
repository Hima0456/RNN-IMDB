```python
from tensorflow.keras.preprocessing.text import one_hot

## word Embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential

import numpy as np
```

```python
### sentences
sent=[  'Hima eats mango',
      'mango on tree',
      'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good girl',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
vocab_size=10000
```

```python
### One Hot Representation
one_hot_repr=[one_hot(words,vocab_size)for words in sent]
one_hot_repr
```

```python
sent_length=9

embedded_docs= pad_sequences(one_hot_repr, padding='post', maxlen=sent_length)
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_length))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
